```python
new_column_names = []                                      #Renaming the columns
for i in data.columns:
    new_column_names.append(i.capitalize())
```

```python
# Column transformation
```

```python
data["Show_id"]
```

```python
data["Show_id"] = data["Show_id"].str.lstrip("s")           #Removing 's' from Show_id
```

```python
data.head()
```

```python
type(data["Show_id"][0])
```

```python
data["Show_id"] = data["Show_id"].astype(int)
```

```python
type(data["Show_id"][0])
```

```python
data["Date_added"] = data["Date_added"].apply(lambda x: x.replace("/","-"))    
#to replace "/" with "-"
```
